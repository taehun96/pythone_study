```python
# 파이썬 표준 라이브러리
# 파이썬이 설치될 때 기본적으로 포함되어 있는 모듈들의 집합

# math : 수학적 연산에 사용되는 모듈
import math

# 1. 올림/내림
# rount : 반올림 - 파이썬 내장 함수
round(3.141592, 2)

# ceil : 올림, 소수점 지정 X
math.ceil(3.14)

# floor :내림, 소수점 지정 X
math.floor(3.14)

# 2. 제곱, 제곱근
# pow(x, y) : 제곱 - x^y를 반환
math.pow(2, 3)

# sqrt(x) : 제곱근 반환
math.sqrt(16)

# 3. 상수 - 괄호없음
# pi : 원주율
math.pi
```

```python
# 실습2. 문제1. 실제 거리 계산
import math
# 두 점 (x1, y1), (x2, y2)의 좌표를 입력받아 두 점 사이의 실제 거리를 소수 둘째 자리까지 출력하세요
x1, x2 = map(int, input("x1, x2을 입력해주세요.").split(","))
y1, y2 = map(int, input("y1, y2을 입력해주세요.").split(","))

# 두 점 사이의 실제 거리를 소수 둘째 자리까지 출력하세요
distance = round(math.sqrt(math.pow(x2 - x1, 2) + math.pow(y2 - y1, 2)), 2)

print(f"두 점 사이의 거리는 {distance}입니다.")
```

```python
# 실습2. 문제2 상품 나누기
import math

students = 18
teachers = 24

# 최대공약수
max_gcd = math.gcd(students, teachers)
# 최소공배수
min_lcm = math.lcm(students, teachers)

print(f"모두 공평하게 나눠 가질 수 있는 최대의 간식 개수 : {max_gcd}")
print(f"포장 단위별로 동시에 맞게 나누어 떨어지는 최소 간식 개수: {min_lcm}")
```

```python
# random 모듈 : 랜덤 값(난수) 생성시 사용
import random

# 1. 난수 생성
# seed(a) : 난수 발생 초기값(seed) 설정 - 최초 생성한 난수를 고정해야 할 때 사용
# random.seed(42)

# random() : 0이상 1미만의 float 난수 반환
random.random()

# uniform(a, b) : a이상 b이하의 실수 난수 반환
random.uniform(1, 10)

# randint(a, b) : a이상 b이하의 정수 난수 반환
random.randint(1, 100)

# randrange(start, stop, step) : 범위 안의 정수 난수 반환, 간격 지정 가능
random.randrange(0, 100, 5)

# 2. 랜덤 선택
fruits = ["apple", "banana", "watermelon", "grape", "orange"]

# choice(seq) : 시퀀스에서 임의의 요소 1개 반환
random.choice(fruits)

# choices(seq, k) : 시퀀스에서 "중복 허용" k개의 요소 리스트를 반환
random.choices(fruits, k=2)

# sample(seq , k) : 시퀀스에서 "중복 없이" k개 요소 리스트를 반환
random.sample(fruits, k=2)

# shuffle() : 시퀀스의 요소를 무작위로 섞음 → 원본 시퀀스를 변경
numbers = [1,2,3,4,5]
random.shuffle(numbers)
numbers
```

```python
# 실습4 가위 바위 보 게임 만들기
import random
rand = ["가위", "바위", "보"]
computer = random.choice(rand)

user = input("가위, 바위, 보 중에 하나를 입력해주세요: ")

while True:
    user = input("가위, 바위, 보 중 하나를 입력하세요: ")
    if (user == "가위") or (user == "바위") or (user == "보"):
        break
    else:
        print("잘못된 입력입니다. 다시 입력해주세요.")

print(f"사용자 : {user}, 컴퓨터 : {computer}")

if user == computer:
    print("비겼습니다.")
elif (user == "보" and computer == "바위") or (user == "가위" and computer == "보") or (user == "주먹" and computer == "가위"):
    print("이겼습니다!")
else:
    print("졌습니다.")
```

```python
# datetime 모듈
# 날짜와 시간의 생성, 조작, 현실 변환과 같은 시간 관련 기능을 제공
import datetime

# 1. 날짜/시간 구하기(datetime의 가장 기본적인 것)
# 현재 날짜와 시간 구하기
now = datetime.datetime.now()

# 오늘 날짜만 구하기
today = datetime.date.today()

# 2. 날짜/시간 형식 변환
# strftime : datetime 객체 → 문자열
# strptime : 문자열 → datetime 객체
formatted = now.strftime("%Y/%m/%d %H:%M:%S")
formatted

parsed = datetime.datetime.strptime(formatted, "%Y/%m/%d %H:%M:%S")
parsed

# 3. 날짜/시간 연산
dt = datetime.date(2025, 7, 7)
# 날짜끼리 연산 가능
passed_time = today - dt
print(f"개강 이후 {passed_time.days}일이 지났습니다.")

# 4. 요일반환 : weekday
# 0: 월요일 ~ 7: 일요일
days = ["월", "화", "수", "목", "금", "토", "일"]
day_num = today.weekday()
days[day_num]

추석 = datetime.date(2025, 10, 6).weekday()
days[추석]

# datetime 또는 date 객체에는 년/월/일 시간 등이 속성으로 들어있음
datetime.datetime.now().year
datetime.date.today().day
```

```python
# 실습5 다음 생일까지 남은 날짜 계산하기
from datetime import datetime

# 사용자 생일 입력
user_birth = input("사용자 생일을 입력하세요 (월-일, 예: 07-25): ")

# 오늘 날짜
today = datetime.today()

# 2025생일
user_birth2025 = datetime.strptime(f"{today.year}-{user_birth}", "%Y-%m-%d")

# 생일 지났을시 2026생일
if user_birth2025 < today:
    user_birth2026 = datetime.strptime(f"{today.year + 1}-{user_birth}", "%Y-%m-%d")
else:
    user_birth2026 = user_birth2025

# 생일 남은 일수
nums_day = (user_birth2026 - today).days

print(f"다음 생일까지 {nums_day}일 남았습니다.")
```

```python
# 실습5 풀이) 다음 생일까지 남은 날짜 계산하기
import datetime

# 사용자로부터 생일을 입력받음
# 달과 일로 쪼갬
birth_month, birth_day = map(int, input("생일을 입력하세요. (예 09/02):").split("/"))
print(birth_month, birth_day)

# 오늘 날짜 구하기(날짜만 있어도 됨)
today = datetime.date.today()

# 올해 생일을 date 객체로 변환(why? 날짜 계산해야 하니까)
birthday_this_year = datetime.date(today.year, birth_month, birth_day)

# 오늘 날짜와 올해 생일을 비교
if today > birthday_this_year:
    # 올해 생일이 지났으면 내년으로 설정
    birthday_next = datetime.date(today.year + 1, birth_month, birth_day)
else:
    # 아니면 올해로 설정
    birthday_next = birthday_this_year

# 남은 일수를 계산
day_left = (birthday_next - today).days

print(f"다음 생일까지 {day_left}일이 남았어요🎂!!")
```

```python
# time 모듈
#시간의 측정, 지연, 변환과 같은 시간 관련 기능 제공

import time
# 1. 시간 반환
# time()
# Unix 타임스탬프로 반환(1970.1.1부터 경과 초)
time.time()

# ctime() : 현재 시간을 문자열로 반환
time.ctime()
time.ctime(0) # 기준시로 반환

# strftime() : 원하는 포맷의 문자열로 시간 객체 변환
lt = time.localtime()
formatted = time.strftime("%Y-%m-%d %H:%M:%S", lt)
print(formatted)

# strptime() : 문자열을 struct__time 객체로 변환
parsed = time.strptime(formatted, "%Y-%m-%d %H:%M:%S")
print(parsed)

# 2. 시간 지연
# sleep(seconds) : 지정한 초만큼 프로그램이 일시 중지
time.sleep(1)
```

```python
# 시간 측정하기
start = time.time()

for i in range(5):
    print(i)
    time.sleep(1)

end = time.time()
print(f"수행시간 : {end-start: .2f}초")
```

```python
# 실습6 타자 연습 게임 만들기
import random
import time

words = ["apple", "banana", "happy","robot", "dream", "green", "mouse", "light", 
        "quick", "jelly", "water", "pencil", "sweet", "cloud", "river", "train",
        "ocean", "bread", "plant", "night", "coffee", "future", "magic", "honey",
        "sugar", "music", "peace", "candy", "tiger", "stone"]

print("[타자 게임] 준비되면 엔터!")
input()
start = time.time()

for i in range(1, 31):
    word = random.choice(words)
    print(f"문제 : {i}")
    print(word)

    while True:
        user_input = input()
        if user_input == word:
            print("통과!")
            break
        else:
            print("오타!")

end = time.time()
print(f"타자시간 : {end-start: .2f}초")
```

```python
# 실습6 풀이) 타자연습게임
import random
import time

words = ["apple", "banana", "happy","robot", "dream", "green", "mouse", "light", 
        "quick", "jelly", "water", "pencil", "sweet", "cloud", "river", "train",
        "ocean", "bread", "plant", "night", "coffee", "future", "magic", "honey",
        "sugar", "music", "peace", "candy", "tiger", "stone"]

n = 1

input("[타자게임] 준비되면 엔터!")

start = time.time()
while n < 11:
    print(f"{n}번 문제")
    question = random.choice(words)
    print(question)
    user_answer = input()

    if user_answer == "종료":
        break

    if question == user_answer:
        print("통과!")
        n += 1
    else:
        print("오타! 다시 도전!")

end = time.time()
et = end - start
print(f" 총 소요시간 : {et: 2f}초")
```
