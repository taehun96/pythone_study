```python
# random ëª¨ë“ˆ : ëœë¤ ê°’(ë‚œìˆ˜) ìƒì„±ì‹œ ì‚¬ìš©
import random

# 1. ë‚œìˆ˜ ìƒì„±
# seed(a) : ë‚œìˆ˜ ë°œìƒ ì´ˆê¸°ê°’(seed) ì„¤ì • - ìµœì´ˆ ìƒì„±í•œ ë‚œìˆ˜ë¥¼ ê³ ì •í•´ì•¼ í•  ë•Œ ì‚¬ìš©
# random.seed(42)

# random() : 0ì´ìƒ 1ë¯¸ë§Œì˜ float ë‚œìˆ˜ ë°˜í™˜
random.random()

# uniform(a, b) : aì´ìƒ bì´í•˜ì˜ ì‹¤ìˆ˜ ë‚œìˆ˜ ë°˜í™˜
random.uniform(1, 10)

# randint(a, b) : aì´ìƒ bì´í•˜ì˜ ì •ìˆ˜ ë‚œìˆ˜ ë°˜í™˜
random.randint(1, 100)

# randrange(start, stop, step) : ë²”ìœ„ ì•ˆì˜ ì •ìˆ˜ ë‚œìˆ˜ ë°˜í™˜, ê°„ê²© ì§€ì • ê°€ëŠ¥
random.randrange(0, 100, 5)

# 2. ëœë¤ ì„ íƒ
fruits = ["apple", "banana", "watermelon", "grape", "orange"]

# choice(seq) : ì‹œí€€ìŠ¤ì—ì„œ ì„ì˜ì˜ ìš”ì†Œ 1ê°œ ë°˜í™˜
random.choice(fruits)

# choices(seq, k) : ì‹œí€€ìŠ¤ì—ì„œ "ì¤‘ë³µ í—ˆìš©" kê°œì˜ ìš”ì†Œ ë¦¬ìŠ¤íŠ¸ë¥¼ ë°˜í™˜
random.choices(fruits, k=2)

# sample(seq , k) : ì‹œí€€ìŠ¤ì—ì„œ "ì¤‘ë³µ ì—†ì´" kê°œ ìš”ì†Œ ë¦¬ìŠ¤íŠ¸ë¥¼ ë°˜í™˜
random.sample(fruits, k=2)

# shuffle() : ì‹œí€€ìŠ¤ì˜ ìš”ì†Œë¥¼ ë¬´ì‘ìœ„ë¡œ ì„ìŒ â†’ ì›ë³¸ ì‹œí€€ìŠ¤ë¥¼ ë³€ê²½
numbers = [1,2,3,4,5]
random.shuffle(numbers)
numbers
```

```python
# ì‹¤ìŠµ5 ë‹¤ìŒ ìƒì¼ê¹Œì§€ ë‚¨ì€ ë‚ ì§œ ê³„ì‚°í•˜ê¸°
from datetime import datetime

# ì‚¬ìš©ì ìƒì¼ ì…ë ¥
user_birth = input("ì‚¬ìš©ì ìƒì¼ì„ ì…ë ¥í•˜ì„¸ìš” (ì›”-ì¼, ì˜ˆ: 07-25): ")

# ì˜¤ëŠ˜ ë‚ ì§œ
today = datetime.today()

# 2025ìƒì¼
user_birth2025 = datetime.strptime(f"{today.year}-{user_birth}", "%Y-%m-%d")

# ìƒì¼ ì§€ë‚¬ì„ì‹œ 2026ìƒì¼
if user_birth2025 < today:
    user_birth2026 = datetime.strptime(f"{today.year + 1}-{user_birth}", "%Y-%m-%d")
else:
    user_birth2026 = user_birth2025

# ìƒì¼ ë‚¨ì€ ì¼ìˆ˜
nums_day = (user_birth2026 - today).days

print(f"ë‹¤ìŒ ìƒì¼ê¹Œì§€ {nums_day}ì¼ ë‚¨ì•˜ìŠµë‹ˆë‹¤.")
```

```python
# ì‹¤ìŠµ5 í’€ì´) ë‹¤ìŒ ìƒì¼ê¹Œì§€ ë‚¨ì€ ë‚ ì§œ ê³„ì‚°í•˜ê¸°
import datetime

# ì‚¬ìš©ìë¡œë¶€í„° ìƒì¼ì„ ì…ë ¥ë°›ìŒ
# ë‹¬ê³¼ ì¼ë¡œ ìª¼ê°¬
birth_month, birth_day = map(int, input("ìƒì¼ì„ ì…ë ¥í•˜ì„¸ìš”. (ì˜ˆ 09/02):").split("/"))
print(birth_month, birth_day)

# ì˜¤ëŠ˜ ë‚ ì§œ êµ¬í•˜ê¸°(ë‚ ì§œë§Œ ìˆì–´ë„ ë¨)
today = datetime.date.today()

# ì˜¬í•´ ìƒì¼ì„ date ê°ì²´ë¡œ ë³€í™˜(why? ë‚ ì§œ ê³„ì‚°í•´ì•¼ í•˜ë‹ˆê¹Œ)
birthday_this_year = datetime.date(today.year, birth_month, birth_day)

# ì˜¤ëŠ˜ ë‚ ì§œì™€ ì˜¬í•´ ìƒì¼ì„ ë¹„êµ
if today > birthday_this_year:
    # ì˜¬í•´ ìƒì¼ì´ ì§€ë‚¬ìœ¼ë©´ ë‚´ë…„ìœ¼ë¡œ ì„¤ì •
    birthday_next = datetime.date(today.year + 1, birth_month, birth_day)
else:
    # ì•„ë‹ˆë©´ ì˜¬í•´ë¡œ ì„¤ì •
    birthday_next = birthday_this_year

# ë‚¨ì€ ì¼ìˆ˜ë¥¼ ê³„ì‚°
day_left = (birthday_next - today).days

print(f"ë‹¤ìŒ ìƒì¼ê¹Œì§€ {day_left}ì¼ì´ ë‚¨ì•˜ì–´ìš”ğŸ‚!!")
```
